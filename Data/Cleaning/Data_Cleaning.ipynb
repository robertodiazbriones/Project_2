```python
import pandas as pd
import numpy as np
```

```python
schools_df = pd.read_csv("cleaned_school_data.csv")
schools_df
```

```python
count = schools_df.loc[schools_df['majority_minority']=="No"]
count.count()
```

```python
count = schools_df.loc[(schools_df['eligible_for_frl'] > 66) & (schools_df['majority_minority']=="No")]
count.count()
```

```python
schools_df['% White'] = list(map(lambda x: x[:-1], schools_df['% White'].values))
schools_df['% Black'] = list(map(lambda x: x[:-1], schools_df['% Black'].values))
schools_df['% Hispanic'] = list(map(lambda x: x[:-1], schools_df['% Hispanic'].values))
schools_df['% Asian'] = list(map(lambda x: x[:-1], schools_df['% Asian'].values))
schools_df['% Other'] = list(map(lambda x: x[:-1], schools_df['% Other'].values))
```

```python
schools_df.to_csv("Cleaned_MI_Schools.csv")
```
